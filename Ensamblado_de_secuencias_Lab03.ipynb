```python
def encontrar_superposicion(seq1, seq2): #Overlap
    max_overlap = 0
    overlap_pos = 0

    for i in range(1,  min(len(seq1),len(seq2))):
        if seq1.endswith(seq2[:i]):
            if i > max_overlap:
                max_overlap = i
                overlap_pos = len(seq1) - i

    return max_overlap, overlap_pos

def complemento_reverso(cadena_adn): #Complemento-Reversa
    complemento = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
    complemento_reverso = [complemento[base] for base in cadena_adn[::-1]]
    return ''.join(complemento_reverso)

def ensamblar_secuencia_greedy(lecturas, index): #Comparar Complemento - Reverso y encontrar "mejor" secuencia de consenso greedy
    secuencia_ensamblada = lecturas[index]
    lecturas_aux = lecturas
    max_overlap, max_overlap_pos = -1,-1
    counter = 0
    visited = [index]
    print(secuencia_ensamblada, end = " ")
    while counter<len(lecturas_aux)-1:
      lecturas_aux = lecturas
      max_overlap, max_overlap_pos = 0,0
      ii = index
      for i in range(len(lecturas_aux)):
          if(i not in visited):
            overlap, overlap_pos = encontrar_superposicion(secuencia_ensamblada, lecturas_aux[i])
            inverted_i = complemento_reverso(lecturas_aux[i])
            overla2, overlap_pos2 = encontrar_superposicion(secuencia_ensamblada, inverted_i)

            if(overlap < overla2):
              overlap = overla2
              lecturas_aux[i] = inverted_i
              overlap_pos = overlap_pos2

            if(overlap >= max_overlap):
              max_overlap = overlap
              max_overlap_pos = overlap_pos
              ii = i


      if max_overlap > 0:
        secuencia_ensamblada += lecturas_aux[ii][max_overlap:]
      else:
        secuencia_ensamblada += lecturas_aux[ii]

      print("->", lecturas_aux[ii], end = " ")
      visited.append(ii)
      counter = counter + 1
    print()
    return secuencia_ensamblada,lecturas_aux

def ensamblar_secuencias(lecturas): #Leer el camino hamiltoniano y ensamblarlo (Indice 0)
    secuencia_ensamblada = lecturas[0]

    for i in range(1, len(lecturas)):
        max_overlap, overlap_pos = encontrar_superposicion(secuencia_ensamblada, lecturas[i])

        if max_overlap > 0:
            secuencia_ensamblada += lecturas[i][max_overlap:]
        else:
            secuencia_ensamblada += lecturas[i]

    return secuencia_ensamblada

def imprimir_secuencias(lecturas, secuencia_original): #Leer el camino hamiltoniano y imprimir su trayectoria
    secuencia_ensamblada = lecturas[0]
    os_len = len(secuencia_original)

    print(lecturas[0],end ="")
    for i in range(os_len - len(lecturas[0])):
      print("-",end ="")
    print()

    for i in range(1, len(lecturas)):
        max_overlap, overlap_pos = encontrar_superposicion(secuencia_ensamblada, lecturas[i])

        if max_overlap > 0:
          secuencia_ensamblada += lecturas[i][max_overlap:]
          for j in range(overlap_pos):
            print("-",end ="")
          print(lecturas[i],end ="")
          for j in range(os_len - len(lecturas[i])-overlap_pos):
            print("-",end ="")
          print()
        else:
          secuencia_ensamblada += lecturas[i]
          for j in range(overlap_pos):
            print("-",end ="")
          print(lecturas[i],end ="")
          for j in range(os_len - len(lecturas[i])-overlap_pos):
            print("-",end ="")
          print()

    return secuencia_ensamblada

def ensamblar_todas_las_secuencias_greedy(lecturas, length): #Ensamblado greedy
    secuencias_ensambladas = []
    len_difference = 9999
    sec = ""
    lecturas_corregidas = [] #Da la mejor lista de secuencias (en caso alguna sea cambiada por su complemento-reverso)
    print("Caminos encontrados:\n")
    for i in range(len(lecturas)):
        secuencia_ensamblada, nuevas_lecturas = ensamblar_secuencia_greedy(lecturas,i)
        secuencias_ensambladas.append(secuencia_ensamblada)
        aux = abs(length-len(secuencia_ensamblada))

        if(aux <= len_difference):
          len_difference = aux
          sec = secuencia_ensamblada
          lecturas_corregidas = nuevas_lecturas

    print("\nThis ->",lecturas_corregidas)
    print("Mejor cadena de longitud:",len(sec))
    print("Mejor cadena de consenso:",sec,"\n")

    return secuencias_ensambladas, lecturas_corregidas

lecturas_adn = ["ATCCGTTGAAGCCGCGGGC", "TTAACTCGAGG","TTAAGTACTGCCCG", "ATCTGTGTCGGG", "CGACTCCCGACACA","CACAGATCCGTTGAAGCCGCGGG","CTCGAGTTAAGTA", "CGCGGGCAGTACTT"]
secuencias_ensambladas, nuevas_lecturas = ensamblar_todas_las_secuencias_greedy(lecturas_adn,55)
print("Secuencias ensambladas greedy:")
for i in range(len(secuencias_ensambladas)):
    print("Iniciando con la cadena ",i+1)
    print(secuencias_ensambladas[i])
print("\nLecturas luego de complemento-reverso optimo: ", nuevas_lecturas)
```

```python
def generar_grafo(lecturas,l):
    grafo = {}

    for nodo1 in lecturas:
        grafo[nodo1] = {}

        for nodo2 in lecturas:
            if nodo2 != nodo1:
                peso_arista, pos_aux = encontrar_superposicion(nodo1, nodo2)
                if peso_arista > l: #Linkage
                    grafo[nodo1][nodo2] = peso_arista
    return grafo

def backtracking(grafo,camino_actual, nodos_visitados):
        if len(camino_actual) == len(nodos):
            return camino_actual
        ultimo_nodo = camino_actual[-1]
        for vecino in grafo[ultimo_nodo]:
            if vecino not in nodos_visitados:
                camino_actual.append(vecino)
                nodos_visitados.add(vecino)
                resultado = backtracking(grafo,camino_actual, nodos_visitados)
                if resultado:
                    return resultado
                nodos_visitados.remove(vecino)
                camino_actual.pop()
        return None

def encontrar_camino_hamiltoniano(grafo):
    nodos = list(grafo.keys())
    for nodo_inicio in nodos:
        camino = [nodo_inicio]
        visitados = set([nodo_inicio])
        resultado = backtracking(grafo,camino, visitados)
        if resultado:
            return resultado

    return None

t_linkage = 0 #Basado en el linkage más débil visto
nodos = nuevas_lecturas
grafo_generado = generar_grafo(nodos, t_linkage)

print("Grafo:")
for nodo, conexiones in grafo_generado.items():
    print(f"Nodo {nodo}: {conexiones}")
print()

# Encontrar un camino Hamiltoniano
camino_hamiltoniano = encontrar_camino_hamiltoniano(grafo_generado)

if camino_hamiltoniano:
    print("Camino Hamiltoniano encontrado:")
    print(" -> ".join(camino_hamiltoniano))
else:
    print("No se encontró un camino Hamiltoniano en el grafo dado.")

#print(camino_hamiltoniano)

secuencia_consenso = ensamblar_secuencias(camino_hamiltoniano)
print("Secuencia de consenso:\n",secuencia_consenso)
print("Tamaño: ", len(secuencia_consenso))
imprimir_secuencias(camino_hamiltoniano,secuencia_consenso)
```
